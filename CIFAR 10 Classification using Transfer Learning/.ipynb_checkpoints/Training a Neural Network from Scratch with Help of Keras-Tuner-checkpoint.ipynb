# Implementing Transfer Learning of CIFAR-10 Dataset

## Part 2: Training a Neural Network from Scratch with Help of Keras-Tuner

### Importing Libraries

```python
import pickle
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Activation
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.model_selection import train_test_split
from keras_tuner.tuners import RandomSearch
from keras_tuner.engine.hyperparameters import HyperParameters
```

### Importing Dataset From Saved Files in Part 1

```python
X = pickle.load(open("features.pickle", "rb"))
y = pickle.load(open("labels.pickle", "rb"))
```

### Visualizing Dataset

```python
labels = pd.read_csv("trainLabels.csv")
cifar_classes = sorted(list(set(labels.label.values)))

f, ax = plt.subplots(1, len(cifar_classes), figsize=(20, 20))

for each_class in cifar_classes:
    index = cifar_classes.index(each_class)
    image = X[y == index][0]
    ax[index].imshow(image)
    ax[index].set_title("Label: {}".format(each_class));

plt.show()
```

### Normalizing Dataset

```python
X = X/255.0
```

### Splitting Dataset into Train and Test Set

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

### Creating Our Neural Network

```python
def build_network(hp):
    network = Sequential()
    
    #Adding initial block to the network
    network.add(Conv2D(hp.Int("conv_input_units", 32, 256, 32), (3, 3), input_shape=X.shape[1:]))
    network.add(Activation('relu'))
    network.add(MaxPooling2D(pool_size=(2, 2)))
    
    #Adding N Convolutional Block in the Network
    for i in range(hp.Int("num_of_conv_layers", 1, 6)):
        network.add(Conv2D(hp.Int(f"conv_{i}_units", 32, 256, 32), (3, 3), input_shape=X.shape[1:]))
        network.add(Activation('relu'))
    
    #Flatten the Layers
    network.add(Flatten())
    
    #Adding Dense Layers
    for i in range(hp.Int("num_of_dense_layers", 1, 6)):
        network.add(Dense(hp.Int(f"dense_{i}_units", 32, 256, 32)))
        network.add(Activation('relu'))
    
    #Output Layer
    network.add(Dense(10))
    network.add(Activation('softmax'))
    
    network.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
    return network
```

```python
tuner = RandomSearch(
    build_network, objective="val_accuracy", max_trials=60,
    executions_per_trial=1, directory="logs/",
    overwrite = False
)

tuner.search(x_train, y_train, epochs=10, validation_data=(x_test, y_test))
```

```python
best_hp = tuner.get_best_hyperparameters()[0].values
best_hp
```

### Conclusion
We have trained with 60 different parameters combination values using Keras Tuner. After running the training for 53m 13s, We reached best validation accuracy of 69%.
